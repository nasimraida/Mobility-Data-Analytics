```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("dc-metrics-data-for-scooter.csv")
#print(df)
df = df.drop(['all'])
#print(df)
#print(df.index)
df = df.reset_index()
df = df.drop(columns=['Total MDS Trips', 'Total Distance (miles)'])
df.columns = ['Time Period', 'Median Trip Distance (miles)', 'Median Trip Duration (minutes)', 'Average Trip Distance (miles)', 'Average Trip Duration (minutes)', 'Average Trips per Day', 'Total Distance (miles)', 'Total MDS Trips']
#print(df)

plt.subplots(2, 2, figsize=(16, 14))
#plt.subplots(2, 2, figsize=(10, 8))

#Trip Distance Graph
x = df['Time Period']
#x = range(24)
y1 = df['Median Trip Distance (miles)']
y2 = df['Average Trip Distance (miles)']
plt.subplot(2, 2, 1)
plt.plot(x, y1, label = "Median", color = "#8460bb", marker = 'o', markersize = 4) #purple
plt.plot(x, y2, label = "Average", color = "#6F91FF", marker = 's', markersize = 4) #blue
plt.title('Trip Distance')
plt.xlabel('Time')
plt.ylabel('Trip Distance (miles)')
plt.xticks(rotation = 45)
plt.legend()

#Trip Duration Grap
y3 = df['Median Trip Duration (minutes)']
y4 = df['Average Trip Duration (minutes)']
plt.subplot(2, 2, 2)
plt.plot(x, y3, label = "Median", color = "#ffccff") #pink
plt.plot(x, y4, label = "Average", color = "#ffd700") #yellow
plt.title('Trip Duration')
plt.xlabel('Time')
plt.ylabel('Trip Duration (minutes)')
plt.xticks(rotation = 45)
plt.legend()

#Total
y5 = df['Total Distance (miles)']
plt.subplot(2, 2, 3)
plt.plot(x, y5, label = "Total", color = "#8cc9ff")
plt.title('Total Trip Distance')
plt.xlabel('Time')
plt.ylabel('Trip Distaince (miles)')
plt.xticks(rotation = 90)
#plt.legend()


#MDS (Mobility Data Speciffication)
y6 = df['Total MDS Trips']
plt.subplot(2, 2, 4)
plt.plot(x, y6, label = "MDS", color = "pink")
plt.title('Total MDS Trips')
plt.xlabel('Time')
plt.ylabel('MDS Trips')
plt.xticks(rotation = 90)
#plt.legend()
plt.suptitle("DC Metrics")
plt.subplots_adjust(wspace=0.25, hspace=0.3)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def makegraph(city):
    df = pd.read_csv(city + '-metrics-data-for-scooter.csv')
    df = df.drop(['all'])
    df = df.reset_index()
    df = df.drop(columns=['Total MDS Trips', 'Total Distance (miles)'])
    df.columns = ['Time Period', 'Median Trip Distance (miles)', 'Median Trip Duration (minutes)', 'Average Trip Distance (miles)', 'Average Trip Duration (minutes)', 'Average Trips per Day', 'Total Distance (miles)', 'Total MDS Trips']
    
    plt.subplots(2, 2, figsize=(16, 14))
    #plt.subplots(2, 2, figsize=(10, 8))

    #Trip Distance Graph
    x = df['Time Period']
    y1 = df['Median Trip Distance (miles)']
    y2 = df['Average Trip Distance (miles)']
    plt.subplot(2, 2, 1)
    plt.plot(x, y1, label = "Median", color = "#8460bb", marker = 'o', markersize = 4) #purple
    plt.plot(x, y2, label = "Average", color = "#6F91FF", marker = 's', markersize = 4) #blue
    plt.title('Trip Distance')
    plt.xlabel('Time')
    plt.ylabel('Trip Distance (miles)')
    plt.xticks(rotation = 45)
    plt.legend()

    #Trip Duration Grap
    y3 = df['Median Trip Duration (minutes)']
    y4 = df['Average Trip Duration (minutes)']
    plt.subplot(2, 2, 2)
    plt.plot(x, y3, label = "Median", color = "#ffccff") #pink
    plt.plot(x, y4, label = "Average", color = "#ffd700") #yellow
    plt.title('Trip Duration')
    plt.xlabel('Time')
    plt.ylabel('Trip Duration (minutes)')
    plt.xticks(rotation = 45)
    plt.legend()

    #Total
    y5 = df['Total Distance (miles)']
    plt.subplot(2, 2, 3)
    plt.plot(x, y5, label = "Total", color = "#8cc9ff")
    plt.title('Total Trip Distance')
    plt.xlabel('Time')
    plt.ylabel('Trip Distaince (miles)')
    plt.xticks(rotation = 90)
    #plt.legend()


    #MDS (Mobility Data Speciffication)
    y6 = df['Total MDS Trips']
    plt.subplot(2, 2, 4)
    plt.plot(x, y6, label = "MDS", color = "pink")
    plt.title('Total MDS Trips')
    plt.xlabel('Time')
    plt.ylabel('MDS Trips')
    plt.xticks(rotation = 90)
    #plt.legend()
    plt.suptitle(city + ' metrics')
    plt.subplots_adjust(wspace=0.25, hspace=0.3)
    plt.show()

makegraph('denver')
```
